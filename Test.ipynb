```python
import statsmodels.api as sm
import numpy as np
import FracRegressionEstimation
import pandas as pd
```

```python
# generate test data 
np.random.seed(0)
n = 1000
X = np.random.rand(n, 3) # 3 features
beta_true = np.array([[0.5], [1.0], [-0.5]])  # true coefficients
y = np.maximum(np.minimum(1 / (1 + np.exp(-X @ beta_true))  + np.random.normal(0, 0.4, size=(n, 1)),1),0)


# create an instance of the estimator and fit the model
estimator = FracRegressionEstimation().fit(X,y)
estimator.beta_
```

```python
pd.DataFrame(y).describe()
```

```python
logistica = sm.Logit(y,X)
logistica.fit().summary() # è proprio la Fractional che fa schifo..... algoritmo converge a quanto fa il 
                          # pacchetto
```
